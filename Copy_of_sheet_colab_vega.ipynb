```
import gspread
import pandas as pd
from google.colab import auth
auth.authenticate_user()
from google.auth import default

creds, _ = default()
gc = gspread.authorize(creds)


worksheet = gc.open('test1').sheet1
# rows = worksheet.get_all_values()

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
# print(rows)

#import pandas as pd
cars2 = pd.DataFrame.from_records(rows) #.rename_axis(index=0, columns=0)
cars2.head()  # display the first five rows
```

```
#cars2.index = cars2.index
#['Name','Miles_per_Gallon','Cylinders','Displacement','Horsepower','Weight_in_lbs','Acceleration','Year','Origin']
header_row = 0
cars2.columns = cars2.iloc[header_row]
cars2 = cars2.drop(header_row)
cars2 = cars2.reset_index(drop=True)
cars2
```

```
import altair as alt

line = alt.Chart(cars2).mark_line().encode(
    alt.X('Year'),
    alt.Y('average(Miles_per_Gallon)')
)

point = alt.Chart(cars2).mark_circle().encode(
    alt.X('Year'),
    alt.Y('average(Miles_per_Gallon)')
)
mark = alt.Chart(cars2).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
).interactive()

line + point
```
